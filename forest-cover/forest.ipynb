```python
import pandas as pd
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import cross_validation
from sklearn.svm import LinearSVC
trainDf = pd.read_csv("train.csv", index_col=0)
testDf = pd.read_csv("test.csv", index_col=0)
%pylab inline
```

```python
trainDf['f2'] = 0
testDf['f2'] = 0
for idx, col in enumerate(trainDf.columns[-42:-2]):
    trainDf['f2'][trainDf[col] == 1] = idx
    testDf['f2'][testDf[col] == 1] = idx
    del trainDf[col]
    del testDf[col]
trainDf['f3'] = 0
testDf['f3'] = 0
for idx, col in enumerate(trainDf.columns[-7:-3]):
    trainDf['f3'][trainDf[col] == 1] = idx
    testDf['f3'][testDf[col] == 1] = idx
    del trainDf[col]
    del testDf[col]
trainDf['f1'] = trainDf.Horizontal_Distance_To_Hydrology ** 2 + trainDf.Vertical_Distance_To_Hydrology ** 2
testDf['f1'] = testDf.Horizontal_Distance_To_Hydrology ** 2 + testDf.Vertical_Distance_To_Hydrology ** 2
```

```python
trainDf.describe()
```

```python
y = trainDf.pop('Cover_Type').values
X = trainDf.values
testX = testDf.values
mod = GradientBoostingClassifier(n_estimators=1500)
scores1 = cross_validation.cross_val_score(mod, X, y, cv=5, n_jobs=-1)
print scores1
```

```python
mod.fit(X, y)
testDf['Cover_Type'] = mod.predict(testX)
```

```python
testDf.Cover_Type.to_csv("submit.csv", header=True)
```
